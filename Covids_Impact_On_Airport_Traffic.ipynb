```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import geopandas as gpd
import plotly.express as px
```

```python
data1 = data.groupby('Country')['PercentOfBaseline'].mean().sort_values(ascending = False).reset_index()
sns.set(font_scale = 1.2)
plt.figure(figsize = [10,7])
sns.barplot(data = data1, x = 'Country', y = 'PercentOfBaseline', palette = 'Blues_d')
plt.ylabel('Percent of baseline')
```

```python
data1 = data.groupby("City")['PercentOfBaseline'].mean().sort_values(ascending = False).reset_index()
sns.set(font_scale = 1.2)
plt.figure(figsize = [10,10])
sns.barplot(data = data1, x = 'PercentOfBaseline', y ='City',palette = 'Blues')
plt.xlabel("Percent of baseline")
```

```python
data1 = data.groupby("State")['PercentOfBaseline'].mean().sort_values(ascending = False).reset_index()
sns.set(font_scale = 1.2)
plt.figure(figsize = [10,10])
sns.barplot(data = data1, x = 'PercentOfBaseline', y = 'State',palette = 'Greens')
plt.xlabel("Percent of baseline")
```

```python
data1 = data.groupby("AirportName")['PercentOfBaseline'].mean().sort_values(ascending = False).reset_index()
sns.set(font_scale =1.2)
plt.figure(figsize = [10,10])
sns.barplot(data = data1, x = 'PercentOfBaseline', y = 'AirportName',palette = 'Blues_d')
plt.xlabel('Percent of baseline')
plt.ylabel("Airport name")
```
